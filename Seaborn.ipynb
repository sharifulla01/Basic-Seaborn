# Count Plot

```python
sns.countplot(x='Gender',data=mart)
```

Create a basic count plot to get number of transaction for each product line and change the figure size by matplotlib

```python

plt.figure(figsize=(15,8))
sns.countplot(x='Product line',data = mart)
```

Add hue to get the count on two categories line Product Line and Gender

```python
plt.figure(figsize=(15,6))
sns.countplot(x="Product line",hue='Gender',data=mart)
```

using different color palette

```python
plt.figure(figsize=(15,6))
sns.countplot(x="Product line",hue='Gender',palette = "BuPu", data=mart)
```

# Basic bar plot

```python
plt.figure(figsize=(16,6))
sns.barplot(x='Product line',y='Total',data=mart)
```

Adding hue to the bar plot

```python
plt.figure(figsize=(16,6))
sns.barplot(x='Product line',y='Total',hue='Gender',data=mart)
```

Make the bar plot horizontal

```python
plt.figure(figsize=(10,6))
sns.barplot(x='Total',y='Product line',hue='Gender',data=mart)
```

Plot the bar in a given order

```python
x=mart['Product line'].sort_values()
x.unique()
plt.figure(figsize=(16,6))
sns.barplot(x='Product line',y='Total',hue='Gender',order=x,data=mart)
```

Add CAP on the error bar

```python
x=mart['Product line'].sort_values()
x.unique()
plt.figure(figsize=(16,6))
sns.barplot(x='Product line',y='Total',hue='Gender',order=x,capsize=0.2,data=mart)
```
